# director

```python
nf[nf["director"].isna()]
```

* we want to impute the missing values of director with the most occuring director for the particular type,list_in 
* To calculate that first we have to unnest the listin column value

```python
# unnesting listed_in column 
df1 = nf["listed_in"].str.split(",",expand=True).add_prefix('listed_in')
df2= pd. merge(nf, df1, left_index=True, right_index=True)
df2.drop("listed_in",axis=1,inplace=True)
df3=pd.melt(df2, id_vars=['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added',
       'release_year', 'rating', 'duration', 'description'],value_name ='listed_in').drop("variable",axis=1)
df3.dropna(subset=['listed_in'],inplace=True)
df3
```

```python
df3.info()# we can see the number of rows increased due to the unnesting of listin column i.2 19323
```

```python
# now just check the most frequently occured director according to type and listed_in
mode_df= df3.groupby(["type","listed_in"])["director"].agg(pd.Series.mode).reset_index()
mode_df
```

```python
#imputing director value with the mode of director groupby type and listed_in 
df4=df3
df4 = df4.merge(mode_df, on = ["type","listed_in" ] , how = 'left')
df4["director"] = np.where(df4["director_x"].isna(), df4["director_y"], df4["director_x"])
df4.drop(["director_x","director_y"],axis=1,inplace=True)
df4
```

```python
df4.director.isna().sum()# there is no null values in director now
```

# country

```python
#lets just chck the missing value s of country column 
df4["country"].isna().sum()
```

* To impute the missing values of the country column we need to calculate the mode of the country according to type,director and Listed_in
* for thet we have to unnest the director and country values

```python
# unnesting director column 
df1 = df4["director"].str.split(",",expand=True).add_prefix('director')
dir2= pd. merge(df4, df1, left_index=True, right_index=True)
dir2.drop("director",axis=1,inplace=True)
dir_df3=pd.melt(dir2, id_vars=['show_id', 'type', 'title', 'cast', 'country', 'date_added',
       'release_year', 'rating', 'duration', 'description', 'listed_in'],value_name ='director').drop("variable",axis=1)
dir_df3.dropna(subset=['director'],inplace=True)
dir_df3
```

```python
# unnesting country column
# let's  replace the nan value of the country column with "missing" first so that we can drop the Exta nan value added due to melting 
# after drop of the the extra values we can set them again to nan and reimpute with the value we want
country_df = dir_df3["country"].str.split(",",expand=True).add_prefix('country')
dir_df3.loc[dir_df3["country"].isna(),"country"]="missing"
country_df2= pd. merge(dir_df3, country_df, left_index=True, right_index=True)
country_df2.drop("country",axis=1,inplace=True)
country_df3=pd.melt(country_df2, id_vars=['show_id', 'type', 'title', 'cast', 'date_added', 'release_year',
       'rating', 'duration', 'description', 'listed_in', 'director'],value_name ='country').drop("variable",axis=1)
country_df3
```

```python
# drop Extra Null value
# setting missing values to np.NaN
country_df3.dropna(subset="country",inplace=True)
country_df3.loc[country_df3['country']=="missing","country"]=np.NaN
country_df3
```

```python
# country having highest frequency according to type,listed_in,director
country_mode =  country_df3.groupby(["type","director","listed_in"])["country"].agg(pd.Series.mode).reset_index()
country_mode
```

```python
#imputing the country column NaN value with the mode
country_df1 = country_df3.merge(country_mode, on = ["type","director","listed_in" ] , how = 'left')
country_df1["country"] = np.where(country_df1["country_x"].isna(), country_df1["country_y"], country_df1["country_x"])  
country_df1.drop(["country_x","country_y"],axis=1,inplace=True)
country_df1
```

```python
country_df1['country'].isna().sum()# There is no null values present for country
```

# cast

```python
country_df1['cast'].isna().sum()#checking for missing values in cast column
```

```python
#checking mode for caste based on type,listed_in and country
cast_df= country_df1.groupby(["type","listed_in","country"])["cast"].agg(pd.Series.mode).reset_index()
cast_df
```

```python
# cast Iputation
cast_df1 = country_df1.merge(cast_df, on = ["type","listed_in",'country'] , how = 'left')
cast_df1["cast"] = np.where(cast_df1["cast_x"].isna(), cast_df1["cast_y"], cast_df1["cast_x"])  
cast_df1.drop(["cast_x","cast_y"],axis=1,inplace=True)
cast_df1
```

```python
#unnesting cast
cast_unique = cast_df1["cast"].str.split(",",expand=True).add_prefix('cast')
final_netflix1= pd. merge(cast_df1, cast_unique, left_index=True, right_index=True)
final_netflix1.drop("cast",axis=1,inplace=True)
final_netflix=pd.melt(final_netflix1, id_vars=['show_id', 'type', 'title', 'date_added', 'release_year',
       'rating', 'duration', 'description', 'listed_in', 'director','country'],value_name ='cast').drop("variable",axis=1)
final_netflix.dropna(subset="cast",inplace=True)
final_netflix

```
